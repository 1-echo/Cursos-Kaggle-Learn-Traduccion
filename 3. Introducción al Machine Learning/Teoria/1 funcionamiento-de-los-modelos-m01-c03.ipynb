# Introducción

Comenzaremos con una visión general de cómo funcionan los modelos de *machine learning* y cómo se utilizan. Esto puede parecer básico si ya has trabajado con modelado estadístico o *machine learning* antes. No te preocupes: avanzaremos pronto hacia la construcción de modelos más potentes.

En este curso irás construyendo modelos a medida que avanzamos a través del siguiente escenario:

Tu primo ha ganado millones de dólares especulando con bienes raíces. Te ha ofrecido convertirte en su socio comercial debido a tu interés en la ciencia de datos. Él aportará el capital y tú aportarás los modelos que predigan cuánto valen distintas viviendas.

Le preguntas a tu primo cómo ha predicho los valores inmobiliarios en el pasado, y él responde que simplemente es intuición. Sin embargo, al indagar un poco más, se hace evidente que ha identificado patrones de precios a partir de las casas que ha visto anteriormente, y utiliza esos patrones para hacer predicciones sobre nuevas propiedades que está considerando.

El *machine learning* funciona de la misma manera. Comenzaremos con un modelo llamado **árbol de decisión** (*Decision Tree*). Existen modelos más sofisticados que ofrecen predicciones más precisas, pero los árboles de decisión son fáciles de entender y constituyen el bloque fundamental de algunos de los mejores modelos en ciencia de datos.

Por simplicidad, empezaremos con el árbol de decisión más simple posible.

![First Decision Trees](https://storage.googleapis.com/kaggle-media/learn/images/7tsb5b1.png)

En la figura anterior observamos que el modelo divide la base de datos según si la vivienda tiene más de dos habitaciones. Si el resultado de esta evaluación es positivo, el precio estimado es de aproximadamente **$188.000**; en caso contrario, es decir, si la vivienda tiene dos habitaciones o menos, el precio estimado es de **$178.000**.

Este modelo divide las viviendas en solo dos categorías. El precio predicho para cualquier casa en evaluación es el **precio promedio histórico** de las viviendas que pertenecen a la misma categoría.

Utilizamos los datos para decidir cómo dividir las casas en dos grupos y, posteriormente, para determinar el precio predicho dentro de cada grupo. Este proceso de capturar patrones a partir de los datos se conoce como **ajustar** (*fitting*) o **entrenar** (*training*) el modelo. Los datos utilizados para **ajustar** el modelo se denominan **datos de entrenamiento** (*training data*).

Los detalles sobre cómo se ajusta el modelo (por ejemplo, cómo se realiza la división de los datos) son lo suficientemente complejos como para dejarlos para más adelante. Una vez que el modelo ha sido ajustado, puedes aplicarlo a nuevos datos para **predecir** los precios de viviendas adicionales.

# Mejorando el Árbol de Decisión

¿Cuál de los siguientes dos árboles de decisión es más probable que resulte del ajuste (*fitting*) de los datos de entrenamiento del mercado inmobiliario?

![Primeros Árboles de Decisión](https://storage.googleapis.com/kaggle-media/learn/images/prAjgku.png)

El árbol de decisión de la izquierda (**Árbol de Decisión 1**) probablemente tenga más sentido, ya que refleja la realidad de que las viviendas con más habitaciones tienden a venderse a precios más altos que aquellas con menos habitaciones. La principal limitación de este modelo es que no captura la mayoría de los factores que influyen en el precio de una vivienda, como el número de baños, el tamaño del terreno, la ubicación, entre otros.

Puedes capturar más factores utilizando un árbol con más **divisiones** (*splits*). A estos se les conoce como árboles más **profundos** (*deeper trees*). Un árbol de decisión que también considere el tamaño total del lote de cada vivienda podría verse de la siguiente manera:

![Árbol de Profundidad 2](https://storage.googleapis.com/kaggle-media/learn/images/R3ywQsR.png)

Para predecir el precio de una vivienda, se recorre el árbol de decisión siguiendo siempre el camino que corresponde a las características de esa casa. El precio predicho se encuentra al final del recorrido del árbol. El punto final donde se realiza la predicción se denomina **hoja** (*leaf*).

Las divisiones y los valores en las hojas estarán determinados por los datos, así que es momento de que revises el conjunto de datos con el que trabajarás.
