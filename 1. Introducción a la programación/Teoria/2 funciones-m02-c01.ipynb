# Introducción

En esta lección, aprenderás cómo organizar tu código utilizando funciones. Una **función** es un bloque de código diseñado para realizar una tarea específica. Como verás, las funciones te permiten ejecutar prácticamente el mismo cálculo varias veces sin necesidad de duplicar código.

# Introducción a las funciones: un ejemplo sencillo

Comenzamos con un ejemplo simple de una función. La función `suma_tres()` acepta cualquier número, le suma tres y luego devuelve el resultado.

````python
# Define la función
def suma_tres(var_entrada):
    var_salida = var_entrada + 3
    return var_salida
````

Cada función se compone de dos partes: un **encabezado** (header) y un **cuerpo** (body).

![](https://storage.googleapis.com/kaggle-media/learn/images/gu0AWhK.png)

## Encabezado
El **encabezado** de la función define el nombre de la función y su(s) argumento(s).  
- Todo encabezado de función comienza con la palabra clave `def`, lo que le indica a Python que estamos a punto de definir una función.
- En el ejemplo, el nombre de la función es `suma_tres`.
- En el ejemplo, el argumento es `var_entrada`. El **argumento** es el nombre de la variable que se utilizará como entrada para la función. Siempre está entre paréntesis que aparecen inmediatamente después del nombre de la función. (_Cabe mencionar que una función también puede no tener argumentos, o puede tener múltiples argumentos. Verás algunos ejemplos de esto más adelante en la lección._)  
- Para cada función, los paréntesis que encierran los argumentos de la función deben ser seguidos por dos puntos `:`.

## Cuerpo
El **cuerpo** de la función especifica el trabajo que realiza la función.
- Cada línea de código en el cuerpo de la función debe estar indentada exactamente con cuatro espacios. Puedes hacer esto presionando la barra espaciadora cuatro veces o presionando una vez la tecla "Tab" en tu teclado. (_A medida que aprendas más sobre Python, es posible que necesites indentar tu código con más de cuatro espacios, pero aprenderás más sobre eso más adelante en este curso._)
- La función realiza su trabajo ejecutando todas las líneas indentadas de arriba hacia abajo.  
    - Toma el argumento como entrada, que en el ejemplo es `var_entrada`.  
    - La función crea una nueva variable `var_saluda` con el cálculo `var_salida = var_entrada + 3`.
    - Luego, la última línea de código, llamada la **instrucción return**, simplemente devuelve el valor de `var_salida` como salida de la función.

El bloque de código anterior solo define la función, pero no la ejecuta. Los detalles del cuerpo de la función tendrán más sentido después del siguiente bloque de código, cuando realmente ejecutemos la función.

# Cómo ejecutar (o "llamar") a una función

Cuando ejecutamos una función, también se le puede llamar "invocar" la función.

En el bloque de código a continuación, ejecutamos la función con `10` como valor de entrada. Definimos una nueva variable `nuevo_numero`, que se establece con el valor de salida de la función.

````python
# Ejecutar la función con 10 como entrada
nuevo_numero = suma_tres(10)

# Verificar que el valor sea 13, como se espera
print(nuevo_numero)
````

![](https://storage.googleapis.com/kaggle-media/learn/images/hlUbxQE.png)

Analizando más a detalle, la función realiza lo siguiente:
- `suma_tres(10)` es el valor que obtenemos como salida cuando proporcionamos `10` como valor de `var_entrada` y llamamos a la función `suma_tres()`. Cuando la función se ejecuta, corre todo el código dentro de su cuerpo, de arriba hacia abajo:
  - Primero calcula `var_salida = var_entrada + 3`, lo que establece `var_salida = 13`.  
  - La última línea de código es la instrucción **return**, que devuelve el valor de `var_saluda`, es decir, `13`.
- Al asignar `nuevo_numero = suma_tres(10)`, establecemos que `nuevo_numero = 13`.

**Nota**: Cuando en este tutorial nos referimos de manera informal a la función `suma_tres()`, usamos paréntesis vacíos después del nombre de la función. Esto es consistente con la forma en que normalmente se explican los ejemplos de código en Python, y los paréntesis vacíos dejan claro que nos estamos refiriendo a una función, y no a una variable u otro objeto de Python. Estos paréntesis deben estar siempre vacíos, incluso si la función tiene argumentos.

# Nombres de funciones

En el ejemplo anterior, el nombre de la función fue elegido por ti. Al nombrar tus propias funciones, deberías usar únicamente letras minúsculas, con las palabras separadas por guiones bajos en lugar de espacios.

Poner nombres a las funciones se volverá natural con el tiempo, y es completamente normal que al principio resulte un poco incómodo. La mejor forma de aprender es leyendo y analizando mucho código en Python.


# Un ejemplo más complejo

Ahora que ya entiendes los conceptos básicos, podemos pasar a un ejemplo con un cálculo más largo.

Supongamos que estás ayudando a un amigo a calcular su pago semanal después de impuestos.
- Está en una banda impositiva del 12% (es decir, el 12% de su salario se destina a impuestos, por lo que solo recibe el 88%), y
- Le pagan por hora, a una tarifa de $15 por hora.

La función que se muestra a continuación calcula el pago semanal en función del número de horas trabajadas. Esta función es más compleja que el primer ejemplo, porque tiene más líneas de código y comentarios. Al igual que en el ejemplo anterior, la función tiene un único argumento (`numero_horas`). En el cuerpo de la función:
- Usamos el valor de `numero_horas` para definir el valor de una nueva variable `pago_antes_impuestos`.
- Usamos el valor de `pago_antes_impuestos` para definir el valor de una nueva variable `pago_despues_impuestos`.
- Devolvemos el valor de la variable `pago_despues_impuestos`.


````python
def calcular_pago(numero_horas):
    # Pago antes de impuestos, basado en una tarifa de $15 por hora
    pago_antes_impuestos = numero_horas * 15
    # Pago después de impuestos, considerando un tramo impositivo del 12%
    pago_despues_impuestos = pago_antes_impuestos * (1 - 0.12)
    return pago_despues_impuestos
````

Llamamos a esta función de la misma manera que llamamos a la primera función. El siguiente bloque de código calcula el pago semanal, basándose en haber trabajado 40 horas. (Después de impuestos, el monto es $528).


````python
# Calcular el pago en base a trabajar 40 horas
pago_completo = calcular_pago(40)
print(pago_completo)
````

Para calcular rápidamente el pago en función de una cantidad diferente de horas trabajadas, solo necesitas proporcionar a la función un número distinto. Por ejemplo, supongamos que tu amigo trabaja 32 horas. (En ese caso, recibe $422.40).


````python
pago_final = calcular_pago(32)
print(pago_final)
````

Como escribiste una función, puedes calcular el pago para diferentes cantidades de horas sin tener que reescribir todo el código de los cálculos una y otra vez.

Las funciones pueden ayudarte a evitar errores en tu código y a ahorrar mucho tiempo. En general, al programar, deberías intentar escribir lo menos posible, ya que cada vez que vuelves a escribir un cálculo, es una nueva oportunidad de introducir accidentalmente un error o un typo.

# Alcance de las variables (*scope*)

Las variables definidas dentro del cuerpo de una función no pueden ser accedidas desde fuera de ella. Por ejemplo, el siguiente bloque de código genera un error, porque `pago_despues_impuestos` solo existe dentro de la función.

````python
## Para observar el error por favor elimina el simbolo " # " de la siguiente línea

#print(pago_despues_impuestos)
````

Obtendrás el mismo error si intentas imprimir `pago_antes_impuestos` o `numero_horas`. Por esta razón, si necesitas cualquier información de una función, debes asegurarte de que aparezca en la instrucción **return** al final de la función.

Nos referimos al **alcance** (*scope*) de una variable como la parte del código donde esta es accesible. Las variables definidas _dentro de una función_ (como `pago_despues_impuestos`) tienen un alcance local limitado únicamente a esa función. En cambio, como ya has visto, las variables definidas _fuera de todas las funciones_ (como `pago_final`) tienen un alcance global y pueden ser accedidas desde cualquier parte del código.


# Funciones con múltiples argumentos

Hasta ahora, has aprendido a definir una función con un solo argumento. Para definir una función con múltiples argumentos, solo necesitas agregar más argumentos dentro de los paréntesis en el encabezado de la función y separarlos con una coma.

Hacemos esto con la función `calcular_pago_mas_argumentos()` que se muestra a continuación, la cual calcula el pago semanal en función de tres argumentos:
- `numero_horas`: número de horas trabajadas en una semana
- `salario_por_hora`: el salario por hora (en $/hora)
- `tasa_impositiva`: porcentaje del salario que se descuenta por impuestos


````python
def calcular_pago_mas_argumentos(numero_horas, salario_por_hora , tasa_impositiva):
    # Pago antes de impuestos
    pago_antes_impuestos = numero_horas * salario_por_hora
    # Pago después de impuestos
    pago_despues_impuestos = pago_antes_impuestos * (1 - tasa_impositiva)
    return pago_despues_impuestos
````

Luego, para llamar a la función, necesitas proporcionar un valor para cada entrada, nuevamente separados por comas.

En el bloque de código a continuación, calculamos el pago después de impuestos para alguien que trabaja 40 horas, gana $24 por hora y se encuentra en un tramo impositivo del 22%.


````python
valor_total_pago = calcular_pago_mas_argumentos(40, 24, .22)
print(valor_total_pago)
````

El siguiente bloque de código produce el mismo resultado que cuando ejecutamos `calcular_pago(40)` en las celdas anteriores, porque `salario_por_hora` está establecido en 15 y `tasa_impositiva` está establecido en 12%, lo cual coincide con la forma en que diseñamos la función `calcular_pago()`.


````python
same_pay_fulltime = calcular_pago_mas_argumentos(40, 15, .12)
print(same_pay_fulltime)
````

Dependiendo de cómo planeemos usar esta nueva función `calcular_pago_mas_argumentos()`, puede resultar más útil que la función original `calcular_pago()`, ya que cubre más casos. En lugar de asumir potencialmente de forma incorrecta el salario por hora y el tramo impositivo, la nueva función permite que el usuario especifique los valores correctos. Sin embargo, si estás seguro de que el salario por hora y el tramo impositivo no van a cambiar, la nueva función es simplemente más compleja de lo necesario. En general, al definir funciones, debes considerar cuánta flexibilidad necesitas según tu caso de uso.

# Funciones sin argumentos

Ten en cuenta que es posible definir funciones sin argumentos y que no tengan una instrucción `return`. La función `imprime_hola()` que aparece en el siguiente bloque de código es un ejemplo de ello.


````python
# Definir la función sin argumentos y sin valor de retorno
def imprime_hola():
    print("¡Hola!")
    print("¡Buenos días!")
    
# Llamar a la función
imprime_hola()
````
