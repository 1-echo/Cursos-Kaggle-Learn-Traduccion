# Imprimir un valor (*Printing*)

Una de las tareas más simples (¡y más importantes!) que puedes pedirle a una computadora es que muestre un mensaje en pantalla.

En Python, le pedimos a la computadora que imprima un mensaje escribiendo el comando `print()` y colocando el mensaje dentro de los paréntesis y entre comillas. A continuación, le pedimos a la computadora que imprima el mensaje `Hola, mundo!`.

````python
print("Hola, mundo!")
````

El código está dentro del recuadro (conocido como una **celda de código**), y la respuesta de la computadora (llamada la **salida del código**) se muestra debajo del recuadro. Como puedes ver, la computadora imprimió el mensaje que queríamos.

# Comentarios

Usamos **comentarios** para anotar lo que está haciendo el código. Ayudan a otras personas a entender tu código, y también pueden ser útiles si no has revisado tu propio código en un tiempo. Hasta ahora, el código que hemos escrito es muy corto, pero las anotaciones se vuelven más importantes cuando has escrito mucho código.

Por ejemplo, en la siguiente celda de código, multiplicamos 3 por 2. También agregamos un comentario (`# Multiplicar 3 por 2`) encima del código para describir lo que está haciendo.


````python
# Multiplicar 3 por 2
print(3 * 2)
````

Para indicarle a Python que una línea es un comentario (y no un código Python a ser ejecutado), necesitas escribir el signo de libra (`#`) como el primer carácter de la línea.

Una vez que Python reconoce este símbolo y reconoce que la línea es un comentario, la ignora completamente. Esto es importante, porque al igual que el inglés o el hindi (¡o cualquier otro idioma!), Python es un lenguaje con reglas muy estrictas que deben seguirse. Sin embargo, Python es más estricto que un oyente humano y simplemente generará un error si no puede entender el código.

Podemos ver un ejemplo de esto en la celda de código a continuación. Python genera un error si eliminamos el primer signo, porque el texto en el comentario no es código Python válido, por lo que no puede ser interpretado correctamente.


````python
## Para observar el error por favor elimina el simbolo " # " de la siguiente línea

# Muliplicar 3 por 2
````

# Variables

Hasta ahora, has usado código para hacer un cálculo y mostrar el resultado, pero ese resultado no se guarda en ningún lado. Sin embargo, es posible que quizás quieras (o necesites) guardar el resultado para trabajar con él más tarde. Para esto, necesitarás usar **variables**.

## Creando variables

La siguiente celda de código crea una variable llamada `var_prueba` y le asigna el valor que obtenemos al sumar 5 y 4.

Luego, imprimimos el valor asignado a la variable, que es 9.

````python
# Crea una variable llamada var_prueba y asígnale el valor de 4+5
var_prueba = 4 + 5

# Imprime el resultado almacenado en la variable
print(var_prueba)
````

En general, para trabajar con una variable, primero necesitas seleccionar el nombre que deseas usar. Los nombres de las variables deben ser cortos y descriptivos. También deben cumplir con varios requisitos:
- No pueden tener espacios (por ejemplo, `var prueba` no está permitido)
- Solo pueden incluir letras, números y guiones bajos (por ejemplo, `var_prueba!` no está permitido)
- Deben comenzar con una letra o un guion bajo (por ejemplo, `1_var` no está permitido)

Luego, para crear la variable, necesitas usar `=` para asignarle el valor que deseas que tenga.

Siempre puedes ver el valor asignado a la variable usando `print()` y colocando el nombre de la variable entre paréntesis.

Con el tiempo, aprenderás a elegir buenos nombres para las variables en Python. Es completamente normal que ahora te sientas incómodo, y la mejor forma de aprender es simplemente viendo mucho código en Python.


## Manipulando las variables

Siempre puedes cambiar el valor asignado a una variable sobrescribiendo el valor anterior.

En la siguiente celda de código, cambiamos el valor de `mi_variable` de 3 a 100.


````python
# Asigna el valor de 3 a una nueva variable
mi_variable = 3

# Imprime el valor asignado a mi_variable
print(mi_variable)

# Cambia el valor de la variable a 100
mi_variable = 100

# Imprime el nuevo valor asignado a mi_variable
print(mi_variable)
````

Ten en cuenta que, en general, siempre que defines una variable en una celda de código, todas las celdas de código que sigan también tienen acceso a esa variable. Por ejemplo, usamos la siguiente celda de código para acceder a los valores de `mi_variable` (de la celda de código anterior) y `var_prueba` (de una etapa anterior en este tutorial).


````python
print(mi_variable)
print(var_prueba)
````

La siguiente celda de código le indica a Python que aumente el valor actual de `mi_variable` en 3.

Para hacer esto, necesitamos usar nuevamente `mi_variable =`. Y, al igual que antes, el nuevo valor que queremos asignar a la variable se coloca a la derecha del signo `=`.



````python
# Aumenta el valor en 3
mi_variable = mi_variable+ 3

# Imprime el valor asignado a my_var
print(mi_variable)
````

## Usar múltiples variables

Es común que el código utilice múltiples variables. Esto es especialmente útil cuando tenemos que realizar un cálculo largo con múltiples entradas.

En la siguiente celda de código, calculamos el número de segundos en cuatro años. Este cálculo utiliza cinco entradas.


````python
# Crear variables
num_anios = 4
dias_por_anio = 365 
horas_por_dia = 24
minutos_por_hora = 60
segundos_por_minuto = 60

# Calcular el número de segundos en cuatro años
total_segundos = segundos_por_minuto * minutos_por_hora * horas_por_dia * dias_por_anio * num_anios
print(total_segundos)
````

Como se calculó anteriormente, hay 126144000 segundos en cuatro años.

Cabe mencionar que *es posible* hacer este cálculo sin variables, simplemente con `60 * 60 * 24 * 365 * 4`, pero es mucho más difícil verificar que el cálculo sin variables no tenga algún error, porque no es tan legible. Cuando usamos variables (como `num_anios`, `dias_por_anio`, etc.), podemos llevar un mejor control de cada parte del cálculo, revisar y corregir más fácilmente cualquier error.

Es importante señalar que es especialmente útil usar variables cuando los valores de las entradas pueden cambiar. Por ejemplo, supongamos que queremos mejorar ligeramente nuestra estimación actualizando el valor del número de días en un año de 365 a 365.25, para tener en cuenta los años bisiestos. Entonces, podemos cambiar el valor asignado a `dias_por_anio` sin modificar ninguna de las otras variables y volver a hacer el cálculo.


````python
# Actualizar para incluir los años bisiestos
dias_por_anio = 365.25

# Calcular el número de segundos en cuatro años
total_segundos = segundos_por_minuto * minutos_por_hora * horas_por_dia * dias_por_anio * num_anios
print(total_segundos)
````

**Nota**: Puede que hayas notado el .0 añadido al final del número, lo cual puede parecer innecesario. Esto se debe al hecho de que en el segundo cálculo usamos un número con una parte fraccionaria (365.25), mientras que en el primer cálculo solo multiplicamos números sin parte fraccionaria. Aprenderás más sobre esto en la Lección 3, cuando cubramos los tipos de datos.


## Depuración (*Debugging*)

````python
## Para observar el error por favor elimina el simbolo " # " de la siguiente línea

# print(horas_por_da)
````

Cuando ves un `NameError` como este, es una indicación de que debes revisar cómo has escrito la variable que se menciona como "no definida". Luego, para corregir el error, solo necesitas corregir la ortografía.


````python
print(horas_por_dia)
````
