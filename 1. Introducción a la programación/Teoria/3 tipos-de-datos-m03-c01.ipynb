# Introducción

Cada vez que creas una variable en Python, esta tiene un valor con un tipo de dato asociado. Existen muchos tipos de datos, como enteros, flotantes, booleanos y cadenas de texto, todos los cuales veremos en esta lección. (Este es solo un pequeño subconjunto de los tipos de datos disponibles; también existen diccionarios, conjuntos, listas, tuplas y mucho más).

Los tipos de datos son importantes porque determinan qué tipo de operaciones puedes realizar con ellos. Por ejemplo, puedes dividir dos números flotantes, pero no puedes dividir dos cadenas de texto. Así, `12.0/2.0` tiene sentido, pero `"gato"/"perro"` no.

Para evitar errores, debemos asegurarnos de que las operaciones que realizamos sean compatibles con los tipos de datos que estamos usando.

# Enteros

Los enteros son números sin parte fraccionaria y pueden ser positivos (`1`, `2`, `3`, ...), negativos (`-1`, `-2`, `-3`, ...) o cero (`0`).

En la celda de código a continuación, asignamos a la variable `x` un valor entero. Luego verificamos su tipo de dato con `type()`, pasando únicamente el nombre de la variable entre los paréntesis.


````python
x = 14
print(x)
print(type(x))
````

En la salida anterior, `<class 'int'>` hace referencia al tipo de dato **int** (entero).

# Flotantes

Los flotantes son números que tienen parte fraccionaria. Pueden tener muchos dígitos después del punto decimal.

````python
numero_pi = 3.141592653589793238462643383279502884197169399375105820974944
print(numero_pi)
print(type(numero_pi))
````

También podemos especificar un número flotante usando una fracción.

````python
numero_pi = 22/7
print(numero_pi)
print(type(numero_pi))
````

Una función especialmente útil para trabajar con fracciones es `round()`. Esta permite redondear un número a una cantidad específica de decimales.

````python
# Round to 5 decimal places
numero_pi_redondeado = round(numero_pi, 5)
print(numero_pi_redondeado)
print(type(numero_pi_redondeado))
````

Siempre que escribas un número con punto decimal, Python lo reconoce como un tipo de dato flotante.

Por ejemplo, `1.` (o `1.0`, `1.00`, etc.) será reconocido como un flotante. Esto ocurre incluso cuando, técnicamente, estos números no tienen parte fraccionaria.

````python
y_flotante = 1.
print(y_flotante)
print(type(y_flotante))
````

# Booleanos

Los booleanos representan uno de dos valores posibles: `True` o `False`. En la celda de código a continuación, `z_uno` se asigna a un valor booleano con valor `True`.

````python
z_uno = True
print(z_uno)
print(type(z_uno))
````

A continuación, `z_dos` se asigna a un valor booleano con valor `False`.

````python
z_dos = False
print(z_dos)
print(type(z_dos))
````

Los booleanos se utilizan para representar el valor de verdad de una expresión. Dado que `1 < 2` es una afirmación verdadera, `z_tres` toma el valor `True`.

````python
z_tres = (1 < 2)
print(z_tres)
print(type(z_tres))
````

De manera similar, como `5 < 3` es una afirmación falsa, `z_cuatro` toma el valor `False`.

````python
z_cuatro = (5 < 3)
print(z_cuatro)
print(type(z_cuatro))
````

Podemos invertir el valor de un booleano usando `not`. Así, `not True` equivale a `False`, y `not False` se convierte en `True`.

````python
z_cinco = not z_cuatro
print(z_cinco)
print(type(z_cinco))
````

Los booleanos serán importantes en la próxima lección, cuando aprendas sobre condiciones y sentencias condicionales.

# Cadenas de texto (Strings)

El tipo de dato string es una colección de caracteres (como letras del alfabeto, signos de puntuación, dígitos numéricos o símbolos) contenidos entre comillas. Las cadenas de texto se utilizan comúnmente para representar texto.

````python
w = "Hola, Python!"
print(w)
print(type(w))
````

Puedes obtener la longitud de una cadena de texto usando `len()`. `"Hola, Python!"` tiene una longitud de 13, ya que contiene 13 caracteres, incluidos el espacio, la coma y el signo de exclamación. Ten en cuenta que las comillas no se incluyen al calcular la longitud.

````python
print(len(w))
````

Un tipo especial de cadena de texto es la cadena vacía, que tiene una longitud de cero.

````python
cadena_muy_corta = ""
print(type(cadena_muy_corta))
print(len(cadena_muy_corta))
````

Si colocas un número entre comillas, este pasa a ser del tipo de dato string.

````python
mi_numero = "1.12321"
print(mi_numero)
print(type(mi_numero))
````

Si tenemos una cadena de texto que puede convertirse a un flotante, podemos usar `float()`.

Esto no siempre funciona. Por ejemplo, podemos convertir `"10.43430"` y `"3"` a flotantes, pero no es posible convertir `"Hola, Python!"` a un flotante.

````python
otro_numero = float(mi_numero)
print(otro_numero)
print(type(otro_numero))
````

Así como puedes sumar dos números (ya sean flotantes o enteros), también puedes sumar dos cadenas de texto. El resultado es una cadena más larga que combina las dos cadenas originales mediante concatenación.

````python
nueva_cadena_larga = "abc" + "def"
print(nueva_cadena_larga)
print(type(nueva_cadena_larga))
````

Ten en cuenta que no es posible restar ni dividir dos cadenas de texto. Tampoco puedes multiplicar dos cadenas entre sí, pero sí puedes multiplicar una cadena por un entero. En este caso, el resultado es una nueva cadena que consiste en la cadena original concatenada consigo misma un número determinado de veces.

````python
otra_cadena_especial = "abc" * 3
print(otra_cadena_especial)
print(type(otra_cadena_especial))
````

Ten en cuenta que no puedes multiplicar una cadena de texto por un número flotante. Intentar hacerlo generará un error.

````python
## Para observar el error por favor elimina el simbolo " # " de la siguiente línea

#no_funciona = "abc" * 3.
````

En el error, la “secuencia” es la cadena `"abc"`, y el “non-int of type 'float'” es el número flotante (`3.`). Por lo tanto, el mensaje de error puede reformularse como: “no se puede multiplicar una cadena por un flotante”.
