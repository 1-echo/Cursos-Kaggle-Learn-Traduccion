# Introducción

Ya has visto que, cuando cambias el valor de entrada de una función, a menudo obtienes un resultado diferente. Por ejemplo, considera una función `suma_cinco()` que simplemente suma cinco a cualquier número y devuelve el resultado. Así, `suma_cinco(7)` devolverá 12 (= 7 + 5) y `suma_cinco(8)` devolverá 13 (= 8 + 5). Observa que, sin importar cuál sea la entrada, la acción que realiza la función siempre es la misma: siempre suma cinco.

Sin embargo, puede que necesites una función cuya acción **dependa del valor de entrada**. Por ejemplo, podrías necesitar una función `suma_tres_u_ocho()` que sume tres si la entrada es menor que 10, y sume ocho si la entrada es 10 o mayor. Entonces, `suma_tres_u_ocho(1)` devolverá 4 (= 1 + 3), mientras que `suma_tres_u_ocho(11)` devolverá 19 (= 11 + 8). En este caso, la acción que realiza la función varía según la entrada.

En esta lección aprenderás a usar **condiciones** y **sentencias condicionales** para modificar el comportamiento de tus funciones.

# Condiciones

En programación, las **condiciones** son expresiones que pueden ser `True` o `False`. Existen muchas formas de escribir condiciones en Python, pero algunas de las más comunes consisten simplemente en comparar dos valores distintos. Por ejemplo, puedes comprobar si 2 es mayor que 3.


```python
print(2 > 3)
```

Python identifica esto como `False`, ya que 2 no es mayor que 3.

También puedes usar condiciones para comparar los valores de variables. En la siguiente celda de código, `variable_uno` tiene un valor de 1 y `variable_dos` tiene un valor de 2. En las condiciones, comprobamos si `variable_uno` es menor que 1 (lo cual es `False`) y si `variable_dos` es mayor o igual que `variable_uno` (lo cual es `True`).

```python
variable_uno = 1
variable_dos = 2

print(variable_uno < 1)
print(variable_dos >= variable_uno)
```

# Conditional statements

**Conditional statements** use conditions to modify how your function runs.  They check the value of a condition, and if the condition evaluates to `True`, then a certain block of code is executed.  (Otherwise, if the condition is `False`, then the code is not run.)  

You will see several examples of this in the following sections.

## "if" statements

The simplest type of conditional statement is an "if" statement.  You can see an example of this in the `evaluate_temp()` function below.  The function accepts a body temperature (in Celcius) as input.
- Initially, `message` is set to `"Normal temperature"`.  
- Then, if `temp > 38` is `True` (e.g., the body temperature is greater than 38°C), the message is updated to `"Fever!"`.  Otherwise, if `temp > 38` is False, then the message is not updated.
- Finally, `message` is returned by the function.  

```python
def evaluar_temperatura(temperatura):
    # Configura un mensaje incial
    mensaje = "Temperatura normal"
    # Actualiza el mensaje solo si la temperatuar es mayor a 38
    if temperatura > 38:
        mensaje = "Fiebre!"
    return mensaje
```

En la siguiente celda de código, llamamos a la función con una temperatura de 37 °C. El mensaje es `"Temperatura normal"`, ya que la temperatura es menor que 38 °C (en este caso, `temperatura > 38` se evalúa como `False`).

```python
print(evaluar_temperatura(37))
```

Sin embargo, si la temperatura es de 39 °C, como este valor es mayor que 38 °C, el mensaje se actualiza a `"Fiebre!"`.

```python
print(evaluar_temperatura(39))
```

Observa que hay dos niveles de sangría (o espaciado):
- El primer nivel de sangría se debe a que siempre es necesario indentar el bloque de código que está dentro de una función.
- El segundo nivel de sangría se debe a que también es necesario indentar el bloque de código que pertenece a la sentencia `"if"`. (Como verás, también será necesario indentar los bloques de código de las sentencias `"elif"` y `"else"`).

Ten en cuenta que, como la sentencia `return` no está indentada dentro de la sentencia `"if"`, siempre se ejecuta, independientemente de si `temperatura > 38` es `True` o `False`.

## Sentencias "if ... else"

Podemos usar sentencias `"else"` para ejecutar código cuando una condición es `False`. El código que está bajo la sentencia `"if"` se ejecuta si la condición es `True`, y el código que está bajo `"else"` se ejecuta si la condición es `False`.

```python
def evaluar_temperatura_con_else(temperatura):
    if temperatura > 38:
        mensaje = "Fiebre!"
    else:
        mensaje = "Temperatura normal"
    return mensaje
```

La función `evaluar_temperatura_con_else()` tiene un comportamiento equivalente al de la función `evaluar_temperatura()`.

En la siguiente celda de código, llamamos a esta nueva función con una temperatura de 37 °C. En este caso, `temperatura > 38` se evalúa como `False`, por lo que se ejecuta el código bajo la sentencia `"else"` y se devuelve el mensaje `"Temperatura normal"`.


```python
print(evaluar_temperatura_con_else(37))
```

Al igual que en la función anterior, indentamos los bloques de código que siguen a las sentencias `"if"` y `"else"`.

## Sentencias "if ... elif ... else"

Podemos usar `"elif"` (abreviatura de *"else if"*) para comprobar si varias condiciones pueden ser verdaderas. La función que se muestra a continuación:
- Primero comprueba si `temperatura > 38`. Si esto es verdadero, el mensaje se establece en `"Fiebre!"`.
- Siempre que el mensaje no se haya establecido todavía, la función comprueba luego si `temperatura > 35`. Si esto es verdadero, el mensaje se establece en `"Temperatura_normal"`.
- Después, si aún no se ha establecido ningún mensaje, la sentencia `"else"` garantiza que el mensaje se establezca en `"Temperatura baja"`.

Puedes pensar en `"elif"` como si dijera: *“bien, la condición anterior (por ejemplo, `temperatura > 38`) era falsa, así que veamos si esta nueva condición (por ejemplo, `temperatura > 35`) podría ser verdadera”*.

```python
def evaluar_temperatura_con_elif(temperatura):
    if temperatura > 38:
        mensaje = "Fiebre!"
    elif temperatura > 35:
        mensaje = "Temperatura normal"
    else:
        mensaje = "Temperatura baja"
    return mensaje
```

En la siguiente celda de código, se ejecuta el bloque de código bajo la sentencia `"elif"`, porque `temperatura > 38` es `False` y `temperatura > 35` es `True`. Una vez que este código se ejecuta, la función omite la sentencia `"else"` y devuelve el mensaje.

```python
evaluar_temperatura_con_elif(36)
```

Finalmente, probamos un caso en el que la temperatura es inferior a 35 °C. Dado que las condiciones de las sentencias `"if"` y `"elif"` se evalúan ambas como `False`, se ejecuta el bloque de código que está dentro de la sentencia `"else"`.

```python
evaluar_temperatura_con_elif(34)
```

# Ejemplo – Cálculos

En los ejemplos anteriores, las sentencias condicionales se utilizaron para decidir cómo establecer los valores de las variables. Sin embargo, también puedes usar sentencias condicionales para realizar distintos cálculos.

En este siguiente ejemplo, supongamos que vives en un país con solo dos tramos impositivos. Todas las personas que ganan menos de 12.000 pagan un 25 % en impuestos, y quienes ganan 12.000 o más pagan un 30 %. La función que se muestra a continuación calcula cuánto impuesto se debe pagar.

```python
def calcular_impuestos(ingresos):
    if ingresos < 12000:
        impuesto_a_pagar = 0.25 * ingresos
    else:
        impuesto_a_pagar = 0.30 * ingresos
    return impuesto_a_pagar
```

La siguiente celda de código utiliza la función.

```python
impuestos_ana = calcular_impuestos(9000)
impuestos_bob = calcular_impuestos(15000)

print(impuestos_ana)
print(impuestos_bob)
```

En cada caso, llamamos a la función `calcular_impuestos()` y utilizamos el valor que devuelve para asignarlo a una variable.
- Para `impuestos_ana`, calculamos los impuestos que debe pagar una persona que gana 9.000. En este caso, llamamos a la función `calcular_impuestos()` con `ingresos` establecido en `9000`. Por lo tanto, `ingresos < 12000` es `True` e `impuestos_a_pagar` se establece en `.25 * 9000`. Luego, se devuelve el valor de `impuestos_a_pagar`.
- Para `impuestos_bob`, calculamos los impuestos que debe pagar una persona que gana 15.000. En este caso, llamamos a la función `calcular_impuestos` con `ingresos` establecido en `15000`. Por lo tanto, `ingresos < 12000` es `False` e `impuestos_a_pagar` se establece en `.30 * 15000`. Luego, se devuelve el valor de `impuestos_a_pagar`.

Antes de pasar a otro ejemplo, ¿recuerdas la función `suma_tres_u_ocho()` de la introducción? Esta función recibe un número como entrada y suma tres si el valor es menor que 10; de lo contrario, suma ocho. ¿Puedes deducir cómo escribirías esta función? Una vez que tengas una respuesta, haz clic en el botón **"Show Hidden Code"** que aparece a continuación para ver la solución.

```python
def suma_tres_u_ocho(numero):
    if numero < 10:
        resultado = numero + 3
    else:
        resultado = numero + 8
    return resultado
```

# Ejemplo – Múltiples sentencias "elif"

Hasta ahora, has visto que `"elif"` se utiliza solo una vez dentro de una función. Sin embargo, no existe un límite en la cantidad de sentencias `"elif"` que puedes usar. Por ejemplo, el siguiente bloque de código calcula la dosis de un medicamento (en mililitros) que se debe administrar a un niño, según su peso (en kilogramos).

**Nota**: Esta función no debe utilizarse como consejo médico y representa un medicamento ficticio.


```python
def obtener_dosis(peso):
    # La dosis es de 1.25 ml para cualquier persona con menos de 5.2 kg
    if peso < 5.2:
        dosis = 1.25
    elif peso < 7.9:
        dosis = 2.5
    elif peso < 10.4:
        dosis = 3.75
    elif peso < 15.9:
        dosis = 5
    elif peso < 21.2:
        dosis = 7.5
    # La dosis es de 10 ml para cualquier persona de 21.2 kg o más
    else:
        dosis = 10
    return dosis
```

La siguiente celda de código ejecuta la función. ¡Asegúrate de que el resultado tenga sentido para ti!
- En este caso, la condición del `"if"` es `False`, y todas las condiciones de las sentencias `"elif"` también se evalúan como `False`, hasta llegar a `peso < 15.9`, que resulta ser `True`, por lo que `dosis` se establece en 5.
- Una vez que una sentencia `"elif"` se evalúa como `True` y se ejecuta su bloque de código, la función omite todas las sentencias `"elif"` y `"else"` restantes. Después de omitirlas, lo único que queda es la sentencia `return`, que devuelve el valor de `dosis`.
- ¡El orden de las sentencias `"elif"` sí importa en este caso! Reordenar las condiciones produciría un resultado muy diferente.

```python
print(obtener_dosis(12))
```
