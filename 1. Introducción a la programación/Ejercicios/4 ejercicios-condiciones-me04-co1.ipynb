En esta pregunta, trabajar√°s con una funci√≥n `get_labels()` que recibe informaci√≥n nutricional de un alimento e imprime los sellos de advertencia correspondientes. Esta funci√≥n recibe varias entradas:
- `food_type`: uno de los valores `"solid"` o `"liquid"`  
- `serving_size`: tama√±o de una porci√≥n (si es s√≥lido, en gramos; si es l√≠quido, en mililitros)  
- `calories_per_serving`: calor√≠as por porci√≥n  
- `saturated_fat_g`: gramos de grasa saturada por porci√≥n  
- `trans_fat_g`: gramos de grasa trans por porci√≥n  
- `sodium_mg`: miligramos de sodio por porci√≥n  
- `sugars_g`: gramos de az√∫car por porci√≥n  

Ten en cuenta que parte del c√≥digo puede resultarte desconocido, ya que no se han compartido los detalles de c√≥mo funcionan algunas funciones como `excess_sugar()` o `excess_saturated_fat()`. A grandes rasgos, estas son funciones que devuelven `True` si se considera que el alimento tiene un exceso de az√∫car o de grasa saturada, respectivamente.

Estas funciones se utilizan dentro de la funci√≥n `get_labels()`, y cada vez que se detecta un exceso (ya sea de az√∫car, grasa saturada, grasa trans, sodio o calor√≠as), se imprime el sello de advertencia correspondiente.

**Nota**: Los nombres de las funciones, argumentos y variables deben ser siempre en ingl√©s para que el sistema sea capaz de reconocer el acierto. Reemplaza el valor de 000 por tu respuesta.

````python
# Importa las funciones necesarias para que get_labels funcione
from learntools.intro_to_programming.ex4q5 import (
    excess_sugar,
    excess_saturated_fat,
    excess_trans_fat,
    excess_sodium,
    excess_calories,
)

def get_labels(food_type, serving_size, calories_per_serving,
               saturated_fat_g, trans_fat_g, sodium_mg, sugars_g):
    # Imprime mensajes seg√∫n los resultados obtenidos
    if excess_sugar(sugars_g, calories_per_serving) == True:
        print("EXCESO AZ√öCARES / EXCESS SUGAR")
    if excess_saturated_fat(saturated_fat_g, calories_per_serving) == True:
        print("EXCESO GRASAS SATURADAS / EXCESS SATURATED FAT")
    if excess_trans_fat(trans_fat_g, calories_per_serving) == True:
        print("EXCESO GRASAS TRANS / EXCESS TRANS FAT")
    if excess_sodium(calories_per_serving, sodium_mg) == True:
        print("EXCESO SODIO / EXCESS SODIUM")
    if excess_calories(food_type, calories_per_serving, serving_size) == True:
        print("EXCESO CALOR√çAS / EXCESS CALORIES")
````
