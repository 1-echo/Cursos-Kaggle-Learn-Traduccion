````python
# Carga los datos de la competencia del Titanic
import pandas as pd
titanic_data = pd.read_csv("../input/titanic/train.csv")

# Muestra las primeras cinco filas de los datos
titanic_data.head()
````

- `total` = n√∫mero total de pasajeros que abordaron el barco

- `menores` = n√∫mero de pasajeros menores de 18 a√±os

````python
# Numero de pasajeros total
total = len(titanic_data)
print(total)

# Numero de pasajeros que sobrevivieron
sobrevivieron = (titanic_data.Survived == 1).sum()
print(sobrevivieron)

# Numero de pasajeros menores de 18
menores = (titanic_data.Age < 18).sum()
print(menores)
````
