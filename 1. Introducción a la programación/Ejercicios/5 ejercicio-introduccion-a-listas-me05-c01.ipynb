Supongamos que est√°s haciendo an√°lisis para un sitio web. Necesitas escribir una funci√≥n que devuelva el **crecimiento porcentual** en el n√∫mero total de usuarios con respecto a una cantidad espec√≠fica de a√±os atr√°s.

Tu funci√≥n `percentage_growth()` debe recibir dos argumentos de entrada:

- `numero_usuarios`: una lista de Python con el n√∫mero total de usuarios en cada a√±o. As√≠, `numero_usuarios[0]` corresponde al total de usuarios en el primer a√±o, `numero_usuarios[1]]` al segundo a√±o, y as√≠ sucesivamente. El √∫ltimo elemento de la lista representa el total de usuarios en el a√±o m√°s reciente completado.
- `anios_atras`: el n√∫mero de a√±os hacia atr√°s que se debe considerar para calcular el porcentaje de crecimiento.

Por ejemplo, supongamos que  
`numero_usuarios = [920344, 1043553, 1204334, 1458996, 1503323, 1593432, 1623463, 1843064, 1930992, 2001078]`.

- si `anios_atras = 7`, deber√≠a devolver aproximadamente `0.66`. Esto corresponde a un crecimiento cercano al 66%, calculado como  
  `(2001078 - 1204334) / 1204334`.
